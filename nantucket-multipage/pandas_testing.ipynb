```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data_agg.csv")
print(df.shape)
print(df.head())
```

```python
df["Object"].value_counts()
```

```python
df["Total"] = df.iloc[:, 3:6].sum(axis=1)
df.head()
```

```python
no_inference = df[df["Object"] == "No overlay"]
inference = df[df["Object"] != "No overlay"]
```

```python
no_inference_mean = no_inference[["Frame Read Time", "Inference Time", "Postprocess Time", "Total"]].mean(axis=0)
no_inference_mean = round(no_inference_mean*1000, 2)
no_inference_mean
```

```python
inference_mean = inference[["Frame Read Time", "Inference Time", "Postprocess Time", "Total"]].mean(axis=0)
inference_mean = round(inference_mean*1000, 2)
inference_mean
```

```python
test = pd.concat([inference_mean, no_inference_mean], axis=1, ignore_index=True)
test2 = pd.DataFrame(test)
test3 = test2.transpose()
test3["Type"] = ["Inference", "No Inference"]
test3 = test3[["Type", "Frame Read Time", "Inference Time", "Postprocess Time", "Total"]]
test3
```
